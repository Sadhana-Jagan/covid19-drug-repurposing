```python
import pandas as pd
df = pd.read_csv("DATA-PREPROESSING/all_drug_data_processed.csv")
df = pd.DataFrame(df)

df.head(10)
```

```python
drugs = df["Name"]
drugs_with_indexing = {}
for i,drug in enumerate(drugs):
    drugs_with_indexing[i] = drug.strip().lower()
```

```python
auto_encoder_input = pd.read_csv("GAE-128-8-new.csv")
aei = pd.DataFrame(auto_encoder_input)
aei.shape
```

```python
aei
```

```python
y = aei['Unnamed: 0']
y_drugs = []
for i in y:
    y_drugs.append(drugs_with_indexing[i])
X_ddr_aglo = aei.drop('Unnamed: 0',axis=1)
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch # importing scipy.cluster.hierarchy for dendrogram
dist_mat = sch.linkage(X_ddr_aglo, method = 'complete',metric='euclidean')
dendrogram = sch.dendrogram(dist_mat) # finding the optimal number of clusters using dendrogram
plt.title('Dendrogram') # title of the dendrogram
plt.xlabel('Drugs') # label of the x-axis
plt.ylabel('Euclidean distances') # label of the y-axis
plt.show() # show the dendrogram
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn import metrics
clustering_model = AgglomerativeClustering(n_clusters=10, metric='euclidean', linkage='ward')
clustering_model.fit(X_ddr_aglo)
print(metrics.silhouette_score(X_ddr_aglo,clustering_model.labels_, metric='euclidean'))
```

```python
df_agglo = pd.DataFrame({'node-value':y_drugs, 'cluster-label':clustering_model.labels_})
```

```python
data = dict_ct_num_test_num
```

```python
import numpy as np
sorted_keys = sorted(data.keys())
labels = [f'Cluster-{k+1}' for k in sorted_keys]
clinical_trials = [data[k][0] for k in sorted_keys]
repurposing_candidates = [data[k][1] for k in sorted_keys]

# Normalize to percentages
totals = [sum(data[k][:2]) for k in sorted_keys]
clinical_pct = [c / t * 100 if t > 0 else 0 for c, t in zip(clinical_trials, totals)]
repurpose_pct = [r / t * 100 if t > 0 else 0 for r, t in zip(repurposing_candidates, totals)]

x = np.arange(len(labels))
bar_width = 0.6

fig, ax = plt.subplots(figsize=(12, 6))

# Plot stacked bars
p1 = ax.bar(x, clinical_pct, bar_width, label='Clinical Trial Drugs', color='lightgreen', edgecolor='black')
p2 = ax.bar(x, repurpose_pct, bar_width, bottom=clinical_pct, label='Prospective Candidates for Repurposing', color='cornflowerblue', edgecolor='black')

# Plot red threshold line at 55%
threshold = 55
ax.axhline(y=threshold, color='red', linewidth=2)

# Annotate bars with counts
for i, (c, r) in enumerate(zip(clinical_trials, repurposing_candidates)):
    if c > 0:
        ax.text(x[i], clinical_pct[i] / 2, str(c), ha='center', va='center')
    if r > 0:
        ax.text(x[i], clinical_pct[i] + repurpose_pct[i] / 2, str(r), ha='center', va='center')

# Customizations
ax.set_ylabel('Percentage')
ax.set_ylim(0, 100)
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.set_title('CLUSTER COMPOSITION')
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
